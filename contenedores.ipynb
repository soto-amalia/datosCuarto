```python
# from datetime import datetime
# from pymongo import MongoClient

# mongo = MongoClient(host='localhost', port=27017)

# documento = {
#     'nombre': 'Isaac',
#     'edad': 21,
#     'intereses': ['Python', 'Ciencia de datos', 'Programar'],
#     'fecha': datetime.now()
# }

# db = mongo.datos
# coleccion = db.docum
# coleccion.insert_one(documento)

# print(db.list_collection_names())
# print(coleccion.find_one())

from datetime import datetime
from pymongo import MongoClient
from bson.objectid import ObjectId

# Conexión
mongo = MongoClient("mongodb://localhost:27017/", serverSelectionTimeoutMS=2000)

# Base y colección
db = mongo.datos
coleccion = db.docum

# Inserta un documento de prueba
documentos = [
    {
        'Id': 10,
        'nombre': 'Ana',
        'edad': 31,
        'intereses': ['Ciencia de datos', 'Programar'],
        'fecha': datetime.now()
    },
    {
        'Id': 11,
        'nombre': 'Luis',
        'edad': 27,
        'intereses': ['Machine Learning', 'Python'],
        'fecha': datetime.now()
    },
    {
        'Id': 12,
        'nombre': 'María',
        'edad': 29,
        'intereses': ['IA', 'Big Data', 'Estadística'],
        'fecha': datetime.now()
    }
]

# coleccion.insert_one(documentos[0])
print(db.list_collection_names())
print(coleccion.find_one())
```

```python

# Buscar y mostrar resultados
c = coleccion.find({"nombre": "Isaac"})
for i in c:
    print(i)

# Contar documentos
total = coleccion.count_documents({})
print("Total de documentos:", total)
```

```python
# Insertar todos los documentos
resultado = coleccion.insert_many(documentos)
print("Documentos insertados con _id:", resultado.inserted_ids)

# Consultar por nombre
print("\nBuscar documento con nombre='Ana':")
print(coleccion.find_one({'nombre': 'Ana'}))

# Consultar por ObjectId (usa un ID real de los insertados)
id_ana = resultado.inserted_ids[0]  # primer documento insertado (Ana)
print("\nBuscar por ObjectId:")
print(coleccion.find_one({'_id': ObjectId(id_ana)}))

# Contar documentos totales
total = coleccion.count_documents({})
print("\nTotal de documentos en la colección:", total)

# Listar todos los documentos
print("\nTodos los documentos en la colección:")
for doc in coleccion.find():
    print(doc)

# Filtrar por edad mayor a 28
for d in coleccion.find({'edad': {'$gt': 28}}):
    print(d)

# Actualizar la edad de Ana
coleccion.update_one({'nombre': 'Ana'}, {'$set': {'edad': 32}})

# Eliminar un documento
coleccion.delete_one({'nombre': 'Luis'})

# Ver cuántos quedan
print("Documentos actuales:", coleccion.count_documents({}))

# Buscar por nombre
print("\nBuscar por nombre = 'Ana'")
print(coleccion.find_one({'nombre': 'Ana'}))

# Buscar por _id específico (usa un ObjectId real de tu base)
id_ana = ObjectId("64b8f3e2f1c2b5e4d6a7c8b9")
print("\nBuscar por ObjectId específico:")
print(coleccion.find_one({'_id': id_ana}))

# Contar documentos en la colección
print("\nTotal de documentos:")
print(coleccion.count_documents({}))
```
